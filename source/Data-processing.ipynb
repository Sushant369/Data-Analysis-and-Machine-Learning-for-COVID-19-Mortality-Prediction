```python
df_sorted_loc_date = df_copy_new.sort_values(["location","date"])
#pre filling based on location and date
df_sorted_loc_date["total_tests_per_thousand_new"] = df_sorted_loc_date["total_tests_per_thousand"].fillna(method=
                                                                                                         "ffill")

df_sorted_loc_date["Flag"]=df_sorted_loc_date.apply(lambda x : False if 
                                                    x["total_tests_per_thousand"]==x["total_tests_per_thousand_new"]
                                                    else True,axis=1)

#adding new test cases to total cases aftet prefilling.
df_sorted_loc_date["total_tests_per_thousand_new1"]=df_sorted_loc_date.apply(lambda x : x["total_tests_per_thousand_new"]+
                                                                             x["new_tests_per_thousand"] if x["Flag"]==True
                                                                            else 
                                                                            x["total_tests_per_thousand"],axis=1)

df_sorted_loc_date.drop(["total_tests_per_thousand","total_tests_per_thousand_new","Flag"],inplace=True,axis=1)

df_sorted_loc_date["total_tests_per_thousand"]=df_sorted_loc_date["total_tests_per_thousand_new1"]

df_sorted_loc_date.drop(["total_tests_per_thousand_new1"],inplace=True,axis=1)

print(df_sorted_loc_date.shape)

df_sorted_loc_date["total_tests_per_thousand"].isna()
```

```python
mean_month_total_tests=df.groupby(['month','location'])["total_tests_per_thousand"].mean()
df_mean_month_total_tests=pd.DataFrame(mean_month_total_tests).reset_index()

mean_loc_total_tests=df.groupby(['location'])["total_tests_per_thousand"].mean()
df_mean_loc_total_tests=pd.DataFrame(mean_loc_total_tests).reset_index()

count=0
def get_mean(val):
    if np.isnan(val['total_tests_per_thousand']):
        location_val=val["location"]
        month_val=val["month"]
        total_tests_val=df_mean_month_total_tests[(df_mean_month_total_tests["location"]==location_val) &
                               (df_mean_month_total_tests["month"]==month_val)]["total_tests_per_thousand"].values[0]
        if np.isnan(total_tests_val):
            total_tests_val=df_mean_loc_total_tests[df_mean_loc_total_tests["location"]==
                                                     location_val]["total_tests_per_thousand"].values[0]
        return total_tests_val
    
df['total_tests_per_thousand'] = df.apply(lambda x:get_mean(x),axis=1)

df['total_tests_per_thousand'].isna().sum()

```
